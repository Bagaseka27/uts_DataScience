```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

data = pd.read_csv('sleep_health_and_lifestyle_dataset.csv')
print(data)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

Daily Steps

```python
data["Daily Steps"].unique()
```

```python
data["Daily Steps"].mode()
langkah_harian_mode = data["Daily Steps"].mode()[0]
print("Nilai modus untuk Daily Steps:", langkah_harian_mode)
```

```python
target_kosong = ['N/A', 'Nan', 'Null', 'nan', 'null', '-']
data['Daily Steps'] = data['Daily Steps'].fillna(langkah_harian_mode)
```

```python
data['Daily Steps'].unique()
```

```python
data.isnull().sum()
```

Heart Rate

```python
data['Heart Rate'].unique()
```

```python
rerata_heart_rate = data['Heart Rate'].mean()
heart_rate = math.ceil(rerata_heart_rate)
print("Rata-rata Heart Rate:", heart_rate)
```

```python
target_kosong = ['N/A', 'Nan', 'Null', 'nan', 'null', '-']
data['Heart Rate'] = data['Heart Rate'].fillna(heart_rate)
data['Heart Rate'].unique()
```

```python
plt.boxplot(data['Heart Rate'])
plt.title('Boxplot Heart Rate')
```

```python
data['Heart Rate'] = data['Heart Rate'].replace(to_replace=750, value=70)

print(data['Heart Rate'].unique())
```

```python
plt.boxplot(data['Heart Rate'])
plt.title('Boxplot Heart Rate')
```

```python
data.isnull().sum()
```
